### 2. Solid vs. Liquid Structure
Why does ice take up *more* space than liquid water, unlike most other substances? Describe the atomic reason.

<details>
<summary><strong>Click to see Solution</strong></summary>

**Answer:** The Crystal Lattice has **holes**.

In ice (solid), the water molecules lock into a rigid hexagonal structure that has empty spaces between the molecules. When ice melts, this structure collapses, and the molecules can pack closer together into the empty spaces, making the liquid denser.
</details>

### 3. Evaporation Energy
Why does blowing on hot soup cool it down? Explain in terms of molecular speed.

<details>
<summary><strong>Click to see Solution</strong></summary>

**Answer:**
1. Evaporation happens when the **fastest** molecules escape the liquid surface.
2. Normally, some of these escapees hit air molecules and bounce back into the soup.
3. **Blowing** removes the humid air full of escaping vapor, preventing them from returning.
4. This ensures only the fast ones leave permanently, leaving the slower (colder) ones behind.
</details>

### 4. The Atomic Hypothesis Challenge
Using only the atomic hypothesis, explain why a balloon expands when you heat it.

<details>
<summary><strong>Click to see Solution</strong></summary>

**Answer:**
1. Heat = **Increased Motion**. The gas atoms inside the balloon move faster.
2. Faster atoms hit the rubber walls **harder** and **more often** (Higher Pressure).
3. This increased outward force pushes the rubber walls outward until the tension matches the new pressure.
</details>
