## Section 2-2: Physics Before 1920

This section summarizes the *pre-quantum* model of physics, focused on particles and forces, with a strong emphasis on the electromagnetic field and spectrum.

### üåç The Classical World Picture
- **Space and Time:** Physics takes place in 3D space (Euclid) and time.
- **Particles:** "Things" are particles‚Äîatoms with inertia. If moving, they keep moving unless a force acts.
- **Forces:** Two main types before 1920:
  - **Short-range forces:** e.g. holding atoms in molecules/solids (complex, chemical)
  - **Long-range force:** Gravity ($F \propto 1/r^2$), simple but whose *why* was unknown.
- **Matter:** Everything is made of moving particles; gas pressure is collisions, wind is net flow, heat is atomic motion, sound is density waves.


### üßÆ What Kinds of Particles?
- By 1920, **92 elements (atoms)** discovered; each with unique chemical properties and a special number (the atomic number).
- **Chemical properties** depend only on number of electrons.
- Atoms were pictured as a **massive, positive nucleus** (later known to have protons & neutrons) surrounded by light, negative electrons.

### ‚ö° Gravity vs Electricity
- **Gravity:** Weak, always attractive, acts at all distances.
- **Electricity:** Much, much stronger, with two types of charge:
  - *Plusses repel, minuses repel, but plus-minus attract.*
- Most atoms are neutral overall, so at large distances, electrical forces nearly cancel.
- If you rub off a few negatives or positives, strong electric forces appear‚Äîwhich is the source of static electricity phenomena.
- *Example:* If two grains of sand had a single charge each and were 30 meters apart, the force would be millions of tons if not cancelled!

### üåä The Electric Field: A Better Model
- It's not just "attraction at a distance." A charge *distorts* space: it creates an **electric field**.
- Charges in fields experience force.
- *If you move a charge back and forth, the field takes time to respond, resulting in a delayed force at a distance.*

### üîÑ Magnetism & The Electromagnetic Field
- **Magnetic effects:** Appear when charges are in motion (e.g., a moving comb, a loop of current).
- Electric and magnetic fields are **two aspects of the same thing:** the electromagnetic field.
- A changing electric field *induces* magnetism, and vice versa.

### üì° Electromagnetic Waves & the Spectrum
When a charge is wiggled rapidly it sends out traveling disturbances ‚Äî **electromagnetic waves**.
- **Light is just EM waves in a certain frequency range!**
- Below/above visible light are radio, TV, radar, X-rays, gamma rays, all part of the spectrum.

### üí¨ Reflection
- Feynman describes how, by the early 20th century, nearly all physics phenomena were "reduced" to particles and a few types of forces, with everything else (pressure, heat, sound, even light) emerging from these fundamental ingredients.
- **Electromagnetism** stood out as a profoundly unifying force, linking seemingly unrelated phenomena across the spectrum, from radio to X-rays and beyond.
- This classical picture set the stage for the next revolution in physics: **quantum theory.**
