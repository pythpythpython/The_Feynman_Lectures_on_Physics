## From Days to Hours: The Pendulum
**Galileo's insight:**
> A pendulum always swings back and forth in equal intervals, as long as the swing is kept small.

**How to test it:**
- Count the number of pendulum swings in one "hour" (measured by an hourglass).
- The count is consistentâ€”pendulums are periodic!

**The mechanical clock:**
- Add a mechanism to count the swings and keep them going.
- This is the pendulum clock of our grandfathers.

## The Second
**Convention:**
- If a pendulum oscillates **3600 times** in one hour,
- And there are **24 hours** in a day,
- Then each period of the pendulum is called **one second**.

**Result:**
- We have divided the day into approximately \(10^5\) parts (100,000 seconds/day).

## Beyond Mechanical Clocks: Electronic Oscillators
**The problem:**
> Mechanical pendulums can't be made arbitrarily fast.

**The solution:**
> Build **electronic oscillators**â€”electrical circuits where current swings back and forth, analogous to a pendulum bob.

**Calibration by cascade:**
- Build a series of oscillators, each with a period 10Ã— shorter than the previous one.
- Count how many fast oscillations fit into one slow oscillation.
- This calibrates the fast clock against the slow clock.

## The Oscilloscope: A Microscope for Time
**The problem:**
> When oscillations are faster than a fraction of a second, we can't count them by eye.

**The tool:**
> The **electron-beam oscilloscope** plots electrical current vs. time on a fluorescent screen.

## Measuring Even Shorter Times: A Different Definition
For times shorter than \(10^{-12}\) seconds, we use a **different technique**:

**The method:**
> Measure the **distance** traveled by a moving object during an event.

**Formula:**
$$
\text{Time} = \frac{\text{Distance}}{\text{Speed}}
$$

**Example: Measuring the Lifetime of a \(\pi^0\) Meson**
- A \(\pi^0\) meson is created in a photographic emulsion.
- It travels at nearly the speed of light.
- Under a microscope, we see it leaves a track about \(10^{-7}\) meters long before disintegrating.
- Lifetime: \(\sim 10^{-16}\) seconds.

## The Frontier: Nuclear and Particle Lifetimes
By extending our techniques further, we can infer:
- **Nuclear vibration periods**
- **Lifetimes of strange resonances** (particles discovered in the 1960s):
  - Lifetime: \(\sim 10^{-24}\) seconds
  - This is about the time it would take light to cross a hydrogen nucleus.
