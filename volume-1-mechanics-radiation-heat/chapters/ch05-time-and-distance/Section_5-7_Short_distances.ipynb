## The Gap: Atoms to Nuclei
**Huge scale difference:**
- Atomic diameter: \(10^{-10}\) m
- Nuclear diameter: \(10^{-15}\) m
- **Gap factor: \(10^5\) (100,000 times smaller)**

**Analogy:**
> If an atom were the size of a football stadium, the nucleus would be the size of a marble at the center.

## Step 4: Measuring Nuclear Size
**The challenge:**
> We can't "see" nuclei, even indirectly. We need a different approach.

**The method:**
> Measure the **effective cross section** \(\sigma\) by scattering high-energy particles.

**What is cross section?**
- \(\sigma\) = the apparent target area presented by a nucleus.
- If nuclei are spherical: \(\sigma = \pi r^2\).

**How to measure \(\sigma\):**
1. Fire a beam of high-energy particles at a thin slab of material.
2. Count how many particles are stopped or deflected.
3. Infer the nuclear size from the fraction absorbed.

## The Nuclear Scattering Experiment
**Fig. 5-10:** Imagined view through 1 cm of carbon if only nuclei were visible.

**Setup:**
- Slab of material: 1 cm thick.
- Number of atomic layers: \(\sim 10^8\).
- Nuclei are so small that they don't overlap (from the particle's perspective).

**Key insight:**
> High-energy particles plow through the electron clouds but are stopped only by direct hits on nuclei.

**Measurement:**
- Let \(n_1\) = number of particles entering the slab.
- Let \(n_2\) = number of particles exiting the slab.
- Fraction blocked: \(\frac{n_1 - n_2}{n_1}\).

**Formula:**
$$
\pi r^2 = \sigma = \frac{A}{N} \cdot \frac{n_1 - n_2}{n_1}
$$
where:
- \(A\) = total area of the slab
- \(N\) = number of atoms in the slab

**Result:**
$$
r_{\text{nucleus}} \approx 1 \text{ to } 6 \times 10^{-15} \text{ meters}
$$

## Historical Note: Defining the Meter
**Original definition (1790s):**
$$
1 \text{ meter} = \frac{\pi}{2} \times 10^{-7} \times R_{\text{Earth}}
$$
(i.e., one ten-millionth of the distance from equator to pole)

**Problem:** Not precise or permanent.

**19th-20th century definition:**
> Distance between two scratches on a platinum-iridium bar in Paris.

**Modern definition (post-1960):**
> An agreed-upon number of wavelengths of a chosen spectral line (e.g., krypton-86).

**Current definition (post-1983):**
> The distance light travels in vacuum in \(1/299{,}792{,}458\) of a second.

(This effectively defines the meter in terms of the speed of light \(c\).)

## Relativity: Distance Depends on the Observer
**Key insight:**
> Two observers moving with respect to each other will **not measure the same distances and times**.

**Reason:**
- Distances and times have different magnitudes depending on the **frame of reference**.
- This is the subject of special relativity (covered in a later chapter).

## Quantum Limits: The Heisenberg Uncertainty Principle
**The fundamental limit on position measurements:**
$$
\Delta x \geq \frac{\hbar}{2\Delta p}
$$
where:
- \(\hbar\) = reduced Planck constant (\(\approx 1.05 \times 10^{-34}\) JÂ·s)
- \(\Delta p\) = uncertainty in momentum

**Interpretation:**
> The more precisely you know where something is, the less you know about how fast it's movingâ€”and vice versa.

**Connection to wave nature:**
> This uncertainty is related to the wave-particle duality of matter.

## Quantum Limits: Time-Energy Uncertainty
**The fundamental limit on time measurements:**
$$
\Delta t \geq \frac{\hbar}{2\Delta E}
$$
where:
- \(\Delta E\) = uncertainty in energy

**Interpretation:**
> To know **more precisely when** something happened, you must know **less about what** happened (because your knowledge of energy is less).

**Deep connection:**
> Both position-momentum and time-energy uncertainties are related to the wave nature of matter.
