### 3. The Force Curve
Describe the relationship between two atoms as they get closer together. At what distance do they attract? At what distance do they repel?
<details>
<summary><strong>Click to see Solution</strong></summary>

1.  **Medium Distance:** They **attract** each other slightly (Van der Waals forces, etc.).
2.  **Close Distance:** The attraction increases.
3.  **Very Close (Squeezed):** They violently **repel** each other. 

This implies atoms are not hard balls, but have a "squishy" radius where repulsion kicks in sharply to prevent them from occupying the same space.
</details>

## Part 2: States of Matter (Water)

### 4. Ice vs. Water (Density)
Unlike most substances, water expands when it freezes. Why? Explain using the atomic arrangement.
<details>
<summary><strong>Click to see Solution</strong></summary>

**The Crystal Lattice has "holes".**

In solid ice, the water molecules lock into a rigid hexagonal structure that necessitates empty spaces between the molecules to maintain the pattern. When ice melts, the thermal motion breaks this rigid structure, and the molecules can "fall" into those empty spaces, allowing them to pack tighter together. Thus, liquid water is denser than ice.
</details>

### 5. The Surface of Water
If you look at the surface of a glass of water with a super-microscope, is the surface still? What is happening there?
<details>
<summary><strong>Click to see Solution</strong></summary>

It is a chaotic, dynamic zone. 
1.  Molecules from the liquid are gaining enough energy to break free and fly out (**Evaporation**).
2.  Molecules from the air (vapor) are crashing into the surface and getting stuck (**Condensation**).

Even if the water level looks constant, there is a massive traffic of billions of atoms trading places every second.
</details>

## Part 3: Atomic Processes (Heat & Pressure)

### 6. What is "Heat"?
In the context of the atomic hypothesis, what exactly is "heat" or "thermal energy"?
<details>
<summary><strong>Click to see Solution</strong></summary>

Heat is simply the **kinetic energy of the random jiggling motion** of the particles. 
*   In a gas: How fast the particles fly and hit things.
*   In a solid: How violently the atoms vibrate in their fixed places.
</details>

### 7. The Blowing Soup Mechanics
Why does blowing on a spoonful of hot soup cool it down?
<details>
<summary><strong>Click to see Solution</strong></summary>

1.  **Evaporation cools:** The molecules that escape are the "fastest" ones (highest energy). When they leave, the average energy of the remaining soup drops.
2.  **The Cloud:** Normally, escaping vapor forms a humid cloud over the soup, and many of those particles bounce *back* into the soup, returning their heat.
3.  **The Wind:** Blowing removes this cloud of returning particles. You prevent the "return" of heat, while allowing the "escape" of heat to continue.
</details>

### 8. Defining Pressure
A gas is confined in a box with a piston. Explain what "pressure" is in terms of atomic collisions.
<details>
<summary><strong>Click to see Solution</strong></summary>

Pressure is the result of billions of atoms bombarding the walls of the container (or the piston) every second. Each impact delivers a tiny "impulse" of force. The average of all these billions of tiny pushes results in a steady outward force we call Pressure.
</details>

### 9. Compressing a Gas (Slow vs. Fast)
*   **Scenario A:** You push a piston down *slowly*.
*   **Scenario B:** You push a piston down *fast*.
Why does the gas get hotter in Scenario B?
<details>
<summary><strong>Click to see Solution</strong></summary>

**The Ping-Pong Ball Effect.**
When an atom hits a stationary piston, it bounces off with the same speed. 
But if the piston is **moving inward**, the atom bounces off *faster* than it came in (picking up momentum from the piston). 
Since the atoms are now moving faster, the temperature (kinetic energy) increases.
</details>

## Part 4: Solutions & Precipitation

### 10. Dissolving Salt
Why does a salt crystal dissolve in water? Why don't the salt ions just stay stuck together?
<details>
<summary><strong>Click to see Solution</strong></summary>

The water molecules are jostling around due to thermal motion. Occasionally, they hit an ion on the edge of the salt crystal with enough force to knock it loose. Furthermore, the electrical attraction of the water molecules to the ions helps pull them away.
</details>

### 11. Dynamic Equilibrium (Saturation)
When you put too much salt in water, it stops dissolving and sits at the bottom. Is the salt at the bottom "dead" or inactive?
<details>
<summary><strong>Click to see Solution</strong></summary>

No! It is in **Dynamic Equilibrium**.
The rate of dissolving matches the rate of precipitation.
*   Ions are constantly being knocked off the crystal into the water.
*   Simultaneously, stray ions in the water are bumping into the crystal and getting stuck.
The *net* amount of salt doesn't change, but the specific atoms are constantly cycling.
</details>

## Part 5: Chemical Reactions

### 12. Burning Carbon
What is the process of "burning" at an atomic level (e.g., coal burning in air)?
<details>
<summary><strong>Click to see Solution</strong></summary>

1.  Carbon atoms in the wood are vibrating.
2.  Oxygen atoms from the air hit the carbon.
3.  If the hit is hard enough, the oxygen snaps onto the carbon to form a new unit (CO or CO2).
4.  This "snapping" releases a massive amount of energy (kinetic energy), causing the surrounding atoms to jiggle much faster (Heat/Fire).
</details>

### 13. Why doesn't wood burn spontaneously?
If oxygen and carbon love to snap together and release energy, why doesn't a tree just burst into flames while standing in the air?
<details>
<summary><strong>Click to see Solution</strong></summary>

**Activation Energy.**
Even though the final state (CO2) is lower energy, there is a repulsion barrier to overcome first. The oxygen and carbon must be pushed close enough against their initial repulsion to "snap" into the attractive zone. At room temperature, the impacts aren't hard enough. You need to "light" the fire (add heat) to get the molecules moving fast enough to overcome this barrier.
</details>

### 14. The Mechanics of Smell
How do we smell the scent of violets?
<details>
<summary><strong>Click to see Solution</strong></summary>

Molecules of the flower substance jiggle loose (sublimate/evaporate) and float through the air. They enter our nose and physically interact with receptors.
*   This proves that "scent" is actually physical matter (atoms) traveling from the source to the nose.
*   It also illustrates how distinguishable different molecules are; slightly different shapes trigger different sensations.
</details>

## Part 6: Advanced Thought Experiments

### 15. Brownian Motion
If you look at tiny dust particles (or pollen) in water under a microscope, they are constantly jiggling. Why?
<details>
<summary><strong>Click to see Solution</strong></summary>

This is direct visual proof of atomic motion. The dust particle is being bombarded by billions of water molecules. 
Normally, the hits cancel out. But because the hits are random, sometimes 100 hits come from the left and only 95 from the right. This slight imbalance pushes the dust particle. We can't see the atoms, but we can see the effect of their random bombardment on the dust.
</details>

### 16. The Speed of Scent (Diffusion)
Air molecules move at roughly the speed of sound (very fast!). Yet, if someone opens a bottle of ammonia across the room, it takes several seconds or minutes to smell it. Why implies the molecules aren't flying in a straight line?
<details>
<summary><strong>Click to see Solution</strong></summary>

**Collisions.**
Although an air molecule moves fast, it can't travel far without hitting another air molecule. It follows a jagged, random path (a "Random Walk"). It spends most of its time bouncing back and forth, only slowly drifting across the room.
</details>
