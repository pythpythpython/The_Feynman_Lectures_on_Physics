# 3-3: Biology
## "Everything that animals do, atoms do"
Feynman argues that if you look closely at biology, you just see physics.

### 1. The Physical Machinery of Life
- **Conservation of Energy:** Discovered by Mayer while studying biological heat. Life is not a magic source of energy; it just transforms chemical energy.
- **Nerves:** They are not magical wires. They are thin tubes (membranes) acting like **capacitors**.
  - Step on a stone $\to$ Pressure changes membrane permeability $\to$ Ions flow in $\to$ Voltage drops $\to$ Neighboring point triggers.
  - It's like a row of dominoes falling.

### 2. The Chemical Factory (The Krebs Cycle)
Inside the cell, elaborate chemical reactions happen to release energy.



**Key Concepts:**
- **Enzymes:** Large protein molecules that act as tools. They lower the "activation energy" hill, allowing reactions to happen easily.
- **ATP/GTP:** The "energy currency" of the cell. The transformation from GTP $\to$ GDP releases the kick needed to move a muscle.

### 3. The Blueprint (DNA)
How does the cell know which enzymes to build? 



- **The Structure:** A double helix of sugar and phosphate, connected by 4 bases (A, B, C, D).
- **The Code:** The sequence (e.g., B-A-A-D-C) is the instruction.
- **Replication:** Because A only pairs with B, and C with D, if you split the chain, each half can rebuild its perfect partner. This is how life reproduces.
- **Protein Synthesis:** DNA sends a copy (RNA) to the Ribosome, which builds the protein.
