# 4-3: Kinetic Energy
## The Energy of Motion
Feynman introduces kinetic energy using the classic example of a **pendulum**, where potential energy transforms into kinetic energy and back again.

## The Pendulum: Energy Transformation
When you pull a pendulum bob aside and release it:
- At the **top** (either end): Maximum PE, zero KE (not moving).
- At the **bottom** (center): Zero PE, maximum KE (moving fastest).

**Key Question:** Where does the gravitational potential energy go when the bob is at the bottom?

> It transforms into **kinetic energy**â€”the energy of motion.

This proves that energy can change *form* but the total remains constant.

## Deriving the Formula for Kinetic Energy
Using the same reasoning as before:
- A moving object has energy because it *could* rise to a certain height if converted to PE.
- The kinetic energy at the bottom must equal \(W \times H\), where \(H\) is the height it could reach.

**But how does height depend on velocity?**
- If you throw something straight up with velocity \(V\), it will reach some height \(H(V)\).
- Feynman states (without proving it yet) that:
    $$
    H = \frac{V^2}{2g}
    $$

Substituting into \(KE = WH\):
$$
KE = W \cdot \frac{V^2}{2g} = \frac{WV^2}{2g}
$$

Or, using \(W = mg\):
$$
KE = \frac{1}{2}mv^2
$$

## Kinetic Energy is Universal
Feynman emphasizes:
> "The fact that motion has energy has nothing to do with the fact that we are in a gravitational field."

- Kinetic energy exists whether the object is falling, sliding, or floating in space.
- The formula \(\frac{1}{2}mv^2\) is universal (with corrections at very high speedsâ€”relativity).

## Two Approximations in the Formulas
Feynman notes that both formulas we've seen so far are **approximations**:

1. **\(PE = mgh\):** Breaks down at extreme heights where gravity weakens (far from Earth).
2. **\(KE = \frac{1}{2}mv^2\):** Breaks down at extreme speeds (near the speed of lightâ€”requires relativity correction).

But when we eventually get the **exact** formulas, the law of conservation of energy holds perfectly.
