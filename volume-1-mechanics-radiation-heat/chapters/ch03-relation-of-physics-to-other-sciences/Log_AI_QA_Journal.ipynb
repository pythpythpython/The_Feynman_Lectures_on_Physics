## üß† Key Concept: The Unsolved Problem
**Question:** *Why is turbulence (water in a pipe) so hard to solve if we know the equations?*

**Answer:** 
- The equations (Navier-Stokes) are **non-linear**. 
- This means small changes in the input (a tiny bump in the water) can create massive changes in the output (a giant whirlpool).
- This sensitivity makes "exact solutions" mathematically impossible for most real-world cases.

**Question:** *Did people really believe life had a magic spark?*

**Answer:**
- Yes, for centuries (Vitalism).

- Feynman puts the final nail in the coffin: "Everything animals do, atoms do."
