# 5-1: Motion
## The Birth of Quantitative Physics
Feynman begins Chapter 5 by tracing the origins of modern physics to **Galileo's experiments on motion** 350 years ago.

## Before Galileo: Philosophy, Not Experiments
Until Galileo, the study of motion was dominated by **Aristotelian philosophy**:
- Arguments were based on *thought* rather than *observation*.
- Ancient Greek philosophers (especially Aristotle) made claims about motion that were accepted as "proven" without testing.

**Example of Aristotelian reasoning:**
> "Heavy objects fall faster than light objects because they have more 'desire' to return to the earth."

This sounds plausibleâ€”but is it true?

## Galileo's Revolutionary Approach
**Galileo was skeptical.** Instead of accepting arguments, he **did an experiment**:

1. He rolled a ball down an **inclined trough**.
2. He didn't just *look*â€”he **measured**:
   - **How far** the ball traveled (distance)
   - **How long** it took (time)

This was revolutionary: **Quantitative observation** became the foundation of physics.

## The Problem: How to Measure Time?
In Galileo's era, there were no stopwatches or accurate clocks.

**Galileo's ingenious solution:**
> He used his **pulse** to count off equal intervals of time!

**The Experiment:**
- Release the ball from the top of the inclined track.
- Count: "one â€¦ two â€¦ three â€¦ four â€¦ five â€¦"
- Have a friend mark where the ball is at each count.
- Measure the distances.

**Result:** The marks were at distances proportional to:
$$
1, 4, 9, 16, 25, \ldots
$$
These are the **square numbers**: \(1^2, 2^2, 3^2, 4^2, 5^2, \ldots\)

## Galileo's Law of Falling Bodies
Galileo expressed his observation as:
$$
D \propto t^2
$$
Or, in modern notation:
$$
D = \frac{1}{2}at^2
$$
where \(a\) is the acceleration (constant).

**What this means:**
- Distance grows with the **square** of time.
- The ball acceleratesâ€”it moves faster and faster as it rolls.

This was the first **quantitative law of motion** in history.

## The Two Fundamental Questions of Motion
Feynman summarizes:
> *The study of motion, which is basic to all of physics, treats with the questions:*
> 1. **Where?** (position in space)
> 2. **When?** (position in time)

All of classical mechanicsâ€”and much of modern physicsâ€”is built on answering these two questions with precision.

## ðŸ§ª Interactive: Galileo's Experiment
Let's simulate Galileo's inclined plane experiment!

```
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
g = 9.8  # gravity (m/s^2)
angle = np.pi / 6  # 30-degree incline
a = g * np.sin(angle)  # acceleration down the plane

# Time beats (Galileo's pulse counts)
t = np.array([1, 2, 3, 4, 5, 6, 7, 8])  # seconds

# Distance traveled
D = 0.5 * a * t**2

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Distance vs Time
ax1.plot(t, D, 'o-', markersize=8, linewidth=2, color='blue')
ax1.set_xlabel('Time (pulse beats)', fontsize=12)
ax1.set_ylabel('Distance (meters)', fontsize=12)
ax1.set_title('Galileo\'s Data: Distance vs Time', fontsize=14)
ax1.grid(True, alpha=0.3)

# Plot 2: Distance vs Time^2 (should be linear!)
ax2.plot(t**2, D, 'o-', markersize=8, linewidth=2, color='red')
ax2.set_xlabel('TimeÂ² (pulse beatsÂ²)', fontsize=12)
ax2.set_ylabel('Distance (meters)', fontsize=12)
ax2.set_title('Linear Relationship: D âˆ tÂ²', fontsize=14)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print('Distance at each pulse beat:')
for i, d in enumerate(D, 1):
    print(f'  Beat {i}: {d:.2f} m')
```
