### 1. The Chess Game: Rules vs. Strategy
Feynman distinguishes between *knowing the rules* and *playing the game*.
**Question:** In the context of physics, what corresponds to "the rules" and what corresponds to "the game"? Give a specific example from meteorology (weather).

<details>
<summary><strong>Solution</strong></summary>

- **The Rules:** The fundamental laws of physics (Newton's laws, thermodynamics, fluid dynamics).
- **The Game:** The actual complex phenomena we observe (a hurricane, a rainstorm).
- **Example:** We know exactly how air moves and how water condenses (the rules), but predicting the exact path of a hurricane next week (the game) is nearly impossible due to complexity.
</details>

### 2. The Atomic Hypothesis & Heat
**Question:** Before 1920, heat was often thought of as a fluid. How does the "motion of particles" model explain:
a) What happens when you heat a gas?
b) What "sound" is in this model?

<details>
<summary><strong>Solution</strong></summary>

a) Heating a gas increases the **average kinetic energy** (speed) of the particles. They hit the walls harder, increasing pressure.
b) Sound is a **wave of excess density** (a compression wave) traveling through the crowd of particles.
</details>

### 3. Electrical vs. Gravitational Strength
**Question:** If you held two grains of sand 30 meters apart, and they were stripped of all their electrons (leaving only positive nuclei), the force between them would be roughly 3 million tons. Why don't we feel this force in everyday life?

<details>
<summary><strong>Solution</strong></summary>

Because matter is almost perfectly **neutral**. Every huge positive charge in the nucleus is balanced by an equal negative charge from electrons. The "3 million tons" of repulsion is exactly cancelled by "3 million tons" of attraction to the electrons. Gravity is weak, but it has no negative charge to cancel it out, so it adds up.
</details>

### 4. The Size of the Nucleus
**Question:** An atom is roughly $10^{-8}$ cm across. The nucleus is $10^{-13}$ cm across. If the nucleus were the size of a grape (1 cm), how big would the whole atom be? (Show your rough calculation).

<details>
<summary><strong>Solution</strong></summary>

- Ratio of Atom/Nucleus = $10^{-8} / 10^{-13} = 10^5$ (100,000 times bigger).
- If Nucleus = 1 cm, then Atom = 100,000 cm = 1,000 meters = **1 km**.
- *Visualization:* A grape in the middle of a stadium is too small; it's more like a grape in the middle of a whole city!
</details>

### 5. Why Atoms Don't Collapse (The Quantum Check)
**Question:** Explain in your own words why the **Heisenberg Uncertainty Principle** prevents the electron from sitting right on top of the nucleus.

<details>
<summary><strong>Solution</strong></summary>

If the electron sits on the nucleus, its position uncertainty ($\Delta x$) is near zero. By the rule $\Delta x \Delta p \geq \hbar/2$, its momentum uncertainty ($\Delta p$) must become infinite. Large momentum means large kinetic energy, which makes the electron fly away. It settles at a distance where the electrical attraction balances this "quantum pressure."
</details>

### 8. The "Particle Zoo"
**Question:** In the 1960s, physicists found dozens of new particles like Baryons and Mesons. Why was this considered a "horrible" condition for physics at the time?

<details>
<summary><strong>Solution</strong></summary>

Because science aims for simplicity (amalgamation). Finding 30+ fundamental particles suggests we haven't found the true bottom layer yet. (Later, quarks were discovered to explain this mess).
</details>

### 9. Antimatter
**Question:** What happens when an electron meets a positron?

<details>
<summary><strong>Solution</strong></summary>

They **annihilate**. Their mass is completely converted into energy in the form of gamma rays (photons), following $E=mc^2$.
</details>

### 10. The Scientific Method
**Question:** Feynman says, "The sole test of the validity of any idea is experiment." What does he mean about experiments in different locations (Stockholm vs. Quito)?

<details>
<summary><strong>Solution</strong></summary>

He means that the *laws* of nature should be the same, but the *conditions* might change. If an experiment gives different results in Quito, it doesn't mean physics is broken; it means there's an environmental variable (like Earth's magnetic field or rotation) affecting the result that we need to discover.
</details>
