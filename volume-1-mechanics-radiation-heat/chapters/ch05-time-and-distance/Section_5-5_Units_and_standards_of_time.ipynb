## The Problem: Whose Clock?
We need a **universal standard** for time. But which one should we choose?

**Option 1: The human pulse?**
- Pulses vary wildly from person to person.
- Not a good standard!

**Option 2: A clock?**
- But whose clock?

**Feynman's Swiss story:**
> A Swiss boy wanted all the clocks in his town to ring noon at the same time. Everyone thought it was a marvelous ideaâ€”**so long as all the other clocks rang noon when his did!**

Everyone wants *their* clock to be the standard.

## The Earth: A Shared Clock
**Solution:**
> Fortunately, we all share one clockâ€”**the Earth**.

For a long time, the **rotational period of the Earth** was taken as the basic standard of time.

**Definition of the second (old):**
$$
1 \text{ second} = \frac{1}{86{,}400} \text{ of an average day}

## The Problem with Earth: It's Not Periodic!
As measurements became more precise, scientists discovered:
- **Some days are longer than others.**
- **Some days are shorter.**
- **On average, the Earth's rotation is slowing down** (days are getting longer over centuries).

**Why?**
- Tidal friction from the Moon.
- Redistribution of mass (melting ice, earthquakes, etc.).

**The realization:**
> The "best" clocks are those which agree with each otherâ€”and they disagree with Earth!

## The Solution: Atomic Clocks
**The breakthrough:**
> Use the vibrations of atoms as a clock.

**Why atoms?**
- Atomic vibrations are **insensitive to temperature** and external conditions.
- Every atom of the same type vibrates at **exactly the same frequency**.
- Atomic clocks are available to everyone (in principle).

**Accuracy:**
- Atomic clocks keep time to **1 part in \(10^9\)** (one second in 30 years).
- Modern atomic clocks (as of Feynman's lecture) are even better.

## Norman Ramsey's Hydrogen Clock
**Recent development (early 1960s):**
> Professor Norman Ramsey at Harvard built an atomic clock based on **hydrogen atom vibrations**.

**Expected accuracy:**
- **100 times better** than previous atomic clocks!
- Measurements are ongoing to confirm.

**Feynman's prediction:**
> Since atomic clocks are much more accurate than astronomical time, scientists will soon define the second in terms of an atomic standard.

(Spoiler: This happened in **1967**. The second is now defined by cesium-133 atom oscillations.)

## The Modern Definition of the Second (Post-1967)
**Official SI definition:**
> The second is the duration of **9,192,631,770 periods** of the radiation corresponding to the transition between two hyperfine levels of the ground state of the cesium-133 atom.

**Translation:**
- Count 9,192,631,770 oscillations of a cesium atom.
- That's one second.

**Why cesium-133?**
- Extremely stable.
- Easy to isolate and measure.
- Reproducible anywhere in the universe.

## Why Does This Matter?
**Practical applications:**
- **GPS satellites** require nanosecond precision (atomic clocks).
- **Internet synchronization** relies on atomic time standards.
- **Scientific experiments** (testing relativity, detecting gravitational waves) need extreme precision.

**Philosophical point:**
> Time is not an absolute, pre-existing thing. It's defined by our best clocks. When we build better clocks, we refine our understanding of time itself.
