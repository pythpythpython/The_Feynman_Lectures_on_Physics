# 3-7: How Did It Get That Way?
## The Historical Question
Feynman distinguishes between studying *processes* (how things work now) and *history* (how they evolved).

### 1. Process vs. History
- **Physics:** Typically asks "How does it work?" (Process). We don't ask how the laws of gravity *evolved*; we assume they are timeless.
- **Biology/Geology/Astronomy:** Ask "How did it get here?" (History). Evolution, Mountain formation, Stellar birth.
- **The Convergence:** If we ever find that physical laws *do* change over time, then physics will become a historical science too.

### 2. The Great Unsolved Problem: Turbulence
Feynman highlights a humbling fact:
> *"Nobody in physics has really been able to analyze it mathematically satisfactorily... It is the analysis of circulating or turbulent fluids."*

- **The Problem:** Take a pipe and push water through it fast. How much pressure is needed?
- **The Reality:** We cannot calculate this from first principles! We have to look it up in a table.
- **The Consequence:** This limits our understanding of weather, stars, and the center of the Earth.

### 3. The Glass of Wine
"The whole universe is in a glass of wine."

| Discipline | What it sees in the glass |
| :--- | :--- |
| **Physics** | Twisting liquid, evaporation, light reflections, atoms. |
| **Geology** | The glass itself (distilled rocks), the age of the elements. |
| **Biology** | Fermentation, enzymes, life (bacteria/yeast). |
| **Psychology** | The consciousness watching it, the pleasure of drinking. |

> **Nature does not know these boundaries.** We divide science for convenience, but the glass of wine is one whole thing.
