```

import ipywidgets as widgets
from IPython.display import display, HTML

questions = [
    {
        "q": "What is energy, according to Feynman?",
        "options": ["A physical substance", "A number that stays constant", "A force", "A type of power"],
        "correct": "A number that stays constant"
    },
    {
        "q": "What is the formula for gravitational potential energy near Earth's surface?",
        "options": ["mgh", "Â½mvÂ²", "mv", "Fd"],
        "correct": "mgh"
    },
    {
        "q": "What is the formula for kinetic energy?",
        "options": ["mgh", "Â½mvÂ²", "mv", "Fd"],
        "correct": "Â½mvÂ²"
    },
    {
        "q": "In a frictionless pendulum, what happens at the bottom of the swing?",
        "options": ["PE is maximum", "KE is maximum", "Total energy is zero", "Energy is lost"],
        "correct": "KE is maximum"
    },
    {
        "q": "What assumption did Feynman use to derive PE = mgh?",
        "options": ["F = ma", "Energy is not conserved", "Perpetual motion is impossible", "Gravity varies with height"],
        "correct": "Perpetual motion is impossible"
    },
    {
        "q": "Which form of energy is stored in a compressed spring?",
        "options": ["Kinetic", "Gravitational", "Elastic", "Chemical"],
        "correct": "Elastic"
    },
    {
        "q": "What famous equation relates mass to energy?",
        "options": ["E = mc", "E = mcÂ²", "E = Â½mcÂ²", "E = mÂ²c"],
        "correct": "E = mcÂ²"
    },
    {
        "q": "Why is KE = Â½mvÂ² only an approximation?",
        "options": ["It ignores gravity", "It breaks down at high speeds (relativity)", "It only works for pendulums", "It assumes friction"],
        "correct": "It breaks down at high speeds (relativity)"
    }
]

score = 0
answered_count = 0
total_questions = len(questions)
result_out = widgets.Output()

def run_quiz():
    print("ðŸ§  Chapter 4 Mastery Quiz\n")
    display(result_out)
    
    for i, q in enumerate(questions):
        display(HTML(f"<b>Q{i+1}: {q['q']}</b>"))
        
        def create_handler(btn, correct_answer, other_btns):
            def handler(b):
                global score, answered_count
                for ob in other_btns:
                    ob.disabled = True
                if b.description == correct_answer:
                    b.style.button_color = '#90EE90'
                    b.icon = 'check'
                    score += 1
                else:
                    b.style.button_color = '#FFB6C1'
                    b.icon = 'times'
                    for ob in other_btns:
                        if ob.description == correct_answer:
                            ob.style.button_color = '#90EE90'
                            ob.icon = 'check'
                answered_count += 1
                if answered_count == total_questions:
                    with result_out:
                        from IPython.display import clear_output
                        clear_output()
                        percentage = (score / total_questions) * 100
                        display(HTML(f"<h3>ðŸŽ‰ Quiz Complete!</h3>"))
                        display(HTML(f"<b>Final Score: {score}/{total_questions} ({percentage:.0f}%)</b>"))
                        if percentage >= 80:
                            print("ðŸŒŸ Excellent! You've mastered energy conservation.")
                        else:
                            print("ðŸ“š Good effort! Review the sections you missed.")
            return handler
        
        btns = []
        for opt in q["options"]:
            btn = widgets.Button(description=opt, layout=widgets.Layout(width='300px'))
            btns.append(btn)
        
        for btn in btns:
            btn.on_click(create_handler(btn, q["correct"], btns))
        
        display(widgets.VBox(btns))
        display(HTML("<hr>"))

run_quiz()
```
