# 3-6: Psychology
## The Machinery of the Mind
Feynman approaches psychology with a physicist's skepticism and curiosity. 

### 1. Psychoanalysis is not Science
Feynman is famously blunt here:
> *"Psychoanalysis is not a science: it is at best a medical process, and perhaps even more like witch-doctoring."*

**Why?** 
- Not because it doesn't help people (witch doctors help people too!).
- But because it lacks **experimental verification**. It has theories (spirits/id/ego), but no controlled data on what works and what doesn't.

### 2. The Physical Basis of Memory
The central scientific problem of psychology/neuroscience is:
> *"When an animal learns something... its brain cell must have changed too, if it is made out of atoms. In what way is it different?"*

- **The Atom of Thought:** If you memorize a phone number, *something* physical in your brain structure must change. Is it a chemical? A connection growing? 
- **Current Status:** In the 1960s, this was a total mystery. Today, we know about **synaptic plasticity** (connections getting stronger), but the full "code" of memory is still unsolved.

### 3. Brains vs. Computers
Feynman touches on the analogy between the brain's "interconnecting wires" and computing machines.
- **Complexity:** The brain is so massive that we cannot analyze it wire-by-wire.
- **The Dog Problem:** We are nowhere near understanding humans. *"If we could even figure out how a dog works, we would have gone pretty far."*
- **Reductionism Limit:** Even if we know how a synapse works (physics), predicting human behavior (psychology) is like predicting the weatherâ€”too complex.
