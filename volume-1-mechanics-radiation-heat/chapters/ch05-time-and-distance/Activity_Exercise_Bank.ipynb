## Section 5-3: Short Times
### Exercise 3.1: Pendulum Math
**Given:**
- A pendulum oscillates 3600 times per hour.
- There are 24 hours in a day.

**Calculate:**
1. How many oscillations per day?
2. What is the period of one oscillation (in seconds)?

> *Show your work...*

```python
# Calculate pendulum oscillations
oscillations_per_hour = 3600
hours_per_day = 24

# 1. Oscillations per day
oscillations_per_day = oscillations_per_hour * hours_per_day
print(f"Oscillations per day: {oscillations_per_day:,}")

# 2. Period of one oscillation (in seconds)
seconds_per_hour = 3600
period_seconds = seconds_per_hour / oscillations_per_hour
print(f"Period: {period_seconds} seconds")
```

### Exercise 3.2: The Distance/Speed Method
**Problem:** A Ï€â° meson travels at 99% the speed of light and leaves a track 10â»â· meters long before decaying.

**Given:**
- Speed of light: \(c = 3 \times 10^8\) m/s
- Distance: \(d = 10^{-7}\) m

**Calculate:** How long did the meson live?

> *Formula:* \(t = \frac{d}{v}\)

```python
# Meson lifetime calculation
c = 3e8  # speed of light (m/s)
v = 0.99 * c  # meson speed
d = 1e-7  # distance traveled (m)

t = d / v
print(f"Meson lifetime: {t:.2e} seconds")
```

### Exercise 4.2: Radioactive Dating
**Scenario:** Carbon-14 has a half-life of 5730 years. You find a bone with 25% of its original C-14 remaining.

**Calculate:** How old is the bone?

**Hint:** If 50% remains after 1 half-life, how many half-lives have passed when 25% remains?

```python
import math

# Carbon-14 dating
half_life = 5730  # years
fraction_remaining = 0.25

# Number of half-lives = log(fraction) / log(0.5)
n_half_lives = math.log(fraction_remaining) / math.log(0.5)
age = n_half_lives * half_life

print(f"Number of half-lives: {n_half_lives}")
print(f"Age of bone: {age:.0f} years")
```

### Exercise 6.2: Radar to Venus
**Given:**
- Speed of light: \(c = 3 \times 10^8\) m/s
- Round-trip radar time to Venus: 500 seconds

**Calculate:** Distance to Venus.

**Formula:** \(d = \frac{c \cdot t}{2}\)

```python
# Radar to Venus
c = 3e8  # m/s
t_roundtrip = 500  # seconds

distance = (c * t_roundtrip) / 2
print(f"Distance to Venus: {distance:.2e} meters")
print(f"That's {distance / 1.5e11:.2f} AU (Earth-Sun distances)")
```

### Exercise 7.2: Nuclear Cross Section
**Given:**
- Slab area: \(A = 1 \text{ cm}^2 = 10^{-4} \text{ m}^2\)
- Number of atoms: \(N = 10^{20}\)
- Particles entering: \(n_1 = 10^6\)
- Particles exiting: \(n_2 = 999{,}000\)

**Calculate:** The nuclear radius.

**Formula:** \(\pi r^2 = \frac{A}{N} \cdot \frac{n_1 - n_2}{n_1}\)

```python
import math

# Nuclear cross section
A = 1e-4  # m^2
N = 1e20
n1 = 1e6
n2 = 999000

sigma = (A / N) * ((n1 - n2) / n1)
r = math.sqrt(sigma / math.pi)

print(f"Cross section: {sigma:.2e} mÂ²")
print(f"Nuclear radius: {r:.2e} meters")
print(f"That's {r / 1e-15:.2f} fermis")
```

### Exercise 7.3: Heisenberg Uncertainty
**Given:**
- \(\hbar = 1.05 \times 10^{-34}\) JÂ·s
- You measure an electron's momentum to within \(\Delta p = 10^{-24}\) kgÂ·m/s

**Calculate:** The minimum uncertainty in its position.

**Formula:** \(\Delta x \geq \frac{\hbar}{2\Delta p}\)

```python
# Heisenberg uncertainty
hbar = 1.05e-34  # JÂ·s
delta_p = 1e-24  # kgÂ·m/s

delta_x = hbar / (2 * delta_p)
print(f"Minimum position uncertainty: {delta_x:.2e} meters")
```
