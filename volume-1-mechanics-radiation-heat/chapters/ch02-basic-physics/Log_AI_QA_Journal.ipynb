## üß† Key Concept: The "Why" of Quantum Mechanics
**Question:** *Why exactly does the uncertainty principle prevent the atom from collapsing?*

**Answer:** 
- In classical physics, an electron spirals into the nucleus because it radiates energy.
- In quantum physics, "falling in" means $\Delta x$ (position uncertainty) gets very small.
- The law $\Delta x \Delta p \geq \hbar/2$ demands that as $\Delta x \to 0$, $\Delta p \to \infty$.
- High momentum variability means high **Kinetic Energy**.
- Therefore, squeezing an electron into a nucleus requires giving it infinite energy. It refuses to go! It finds a stable orbit where the electrical pull balances this "quantum refusal" to be confined.

## üß† Key Concept: The Strong Force
**Question:** *If protons repel each other, why doesn't the nucleus explode?*

**Answer:**
- There is a third force involved: The **Strong Nuclear Force**.
- It is much stronger than electricity, but only at short range ($10^{-15}$ m).
- It acts like "nuclear glue" for protons and neutrons.

## üß† Key Concept: Fields vs. Particles
**Question:** *Is light a wave or a particle?*

**Answer:**
- It is **both** (and neither). It is a quantum field excitation.
- At low frequencies (Radio), it behaves like a classical wave.
- At high frequencies (Gamma rays), it behaves like a bullet (particle).
- Quantum mechanics unifies these views.
