## A Different Kind of Clock: Radioactive Decay
**Key insight:**
> Radioactive decay is not periodic (like a pendulum), but it has a **new kind of regularity**.

**The pattern:**
- A radioactive sample decreases by the **same fraction** for each equal time interval.
- If it decreases to **1/2** its original amount in time \(T\) (the "half-life"),
- Then it decreases to **1/4** in another \(T\),
- And to **1/8** in another \(T\), and so on.

**Mathematical formula:**
$$
\text{Fraction remaining after time } t = \left( \frac{1}{2} \right)^{t/T}
$$

## Uranium Dating: Measuring Billions of Years
**For older rocks, use isotopes with longer half-lives:**

**Uranium-238:**
- Half-life: \(~10^9\) years (1 billion years).
- Uranium decays into **lead**.

**The method:**
1. When a rock forms, uranium and lead are chemically separated.
2. Over time, uranium decays into lead.
3. Measure the ratio of uranium to lead today.
4. Calculate how much time has passed.

**Result:**
- The oldest rocks on Earth: **several billion years old**.
- The age of the Earth (using uranium in oceans): **~4.5 billion years**.

## The Age of the Earth and the Universe
**Confirmation:**
> Meteorites that land on Earth also date to **~4.5 billion years**.

**Interpretation:**
- The Earth formed from rocks floating in space.
- Meteorites are leftover material from that same era.

**The Universe:**
- The universe (or at least our part of it) began **10â€“12 billion years ago**.
- Modern cosmology (2025) refines this to **~13.8 billion years**.
