# 4-4: Other forms of energy
## The Full Catalog
Feynman expands the energy inventory beyond gravitational and kinetic to include elastic, heat, electrical, chemical, nuclear, and mass energy.

## Elastic Energy
When you stretch a spring, you do work. The spring can then lift a weight, so it has stored energy.
- **Formula:** \(E_{elastic} = \frac{1}{2}kx^2\) (where \(k\) is the spring constant and \(x\) is displacement).
- **Behavior:** The spring oscillates, converting elastic energy â†” kinetic energy.
- **But eventually it stops.** Where does the energy go?

## Heat Energy: The Hidden Motion
> "We lose track of that energy; we find the atoms are wiggling inside in a random and confused manner."

- When things rub, slide, or compress, atoms jiggle randomly.
- This **internal kinetic energy** is what we call **heat**.
- It is not a new form of energyâ€”it is just kinetic energy we cannot see or control.
- We measure it with thermometers.

## The Full Energy Catalog
| Form | What it is | Example |
| :--- | :--- | :--- |
| **Gravitational** | Position in a gravitational field | \(mgh\) |
| **Kinetic** | Motion | \(\frac{1}{2}mv^2\) |
| **Elastic** | Stretching/compressing | Spring, rubber band |
| **Heat** | Random atomic motion | Friction, temperature |
| **Electrical** | Pushing/pulling by charges | Batteries, capacitors |
| **Radiant** | Electromagnetic waves | Light, radio waves |
| **Chemical** | Attraction between atoms | Burning fuel, digestion |
| **Nuclear** | Arrangement inside nuclei | Fission, fusion |
| **Mass** | Sheer existence | \(E = mc^2\) |

## Mass Energy: \(E = mc^2\)
Feynman introduces Einstein's most famous equation:
> "An object has energy from its sheer existence."

- If a positron and electron meet and annihilate, they release energy as light.
- The amount is \(E = mc^2\), where \(m\) is the mass that disappeared.
- This is the ultimate form of energy: **mass itself is energy**.

## Available Energy vs. Total Energy
> "Although we know for a fact that energy is conserved, the energy available for human utility is not conserved so easily."

- The ocean has enormous heat energy (random atomic motion).
- But we cannot *extract* it without putting in more energy elsewhere.
- This leads to the concept of **entropy** and the Second Law of Thermodynamics.

## Where Do We Get Energy?
- **Sun:** Powers rain, wind, plants (which become coal/oil).
- **Uranium:** Nuclear fission.
- **Hydrogen:** Nuclear fusion (controlled fusion is the dream).

Feynman's challenge:
> "With 150 gallons of running water a minute, you have enough fuel to supply all the energy used in the United States today! Therefore it is up to the physicist to figure out how to liberate us from the need for having energy. It can be done."
