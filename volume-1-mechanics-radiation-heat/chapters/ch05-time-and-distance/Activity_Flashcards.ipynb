# Activity: Flashcards
## Chapter 5: Time and Distance

These flashcards help you memorize key concepts, definitions, and scales. Read the question, try to answer it, then reveal the answer.

---
## Card 1
**Q:** What is Feynman's practical definition of time?

<details>
<summary>Click to reveal answer</summary>

**A:** "Time is how long we wait." He admits we can't define it philosophically, but we can measure it operationally.
</details>

---
## Card 2
**Q:** How many seconds are in one day?

<details>
<summary>Click to reveal answer</summary>

**A:** 86,400 seconds (24 hours × 3600 seconds/hour)
</details>

---
## Card 3
**Q:** What is the modern (post-1967) definition of the second?

<details>
<summary>Click to reveal answer</summary>

**A:** The duration of 9,192,631,770 periods of radiation from a cesium-133 atom transition.
</details>

---
## Card 4
**Q:** What is a fermi?

<details>
<summary>Click to reveal answer</summary>

**A:** 1 fermi = 10⁻¹⁵ meters (the typical scale of atomic nuclei)
</details>

---
## Card 5
**Q:** What is the diameter of a typical atom?

<details>
<summary>Click to reveal answer</summary>

**A:** About 10⁻¹⁰ meters (1 Ångström)
</details>

## Card 6
**Q:** Why can't we use a meter stick to measure the distance between two mountain tops?

<details>
<summary>Click to reveal answer</summary>

**A:** It's impractical—we need a different operational definition. Triangulation (measuring angles) is more practical.
</details>

---
## Card 7
**Q:** What is the distance to the Moon?

<details>
<summary>Click to reveal answer</summary>

**A:** About 4 × 10⁸ meters (400,000 km)
</details>

---
## Card 8
**Q:** How do we measure the distance to nearby stars?

<details>
<summary>Click to reveal answer</summary>

**A:** Triangulation using Earth's orbit as a baseline (measure the star's position 6 months apart).
</details>

---
## Card 9
**Q:** What is the Heisenberg uncertainty principle for position and momentum?

<details>
<summary>Click to reveal answer</summary>

**A:** Δx · Δp ≥ ℏ/2, where ℏ = 1.05 × 10⁻³⁴ J·s
</details>

---
## Card 10
**Q:** What is the wavelength of visible light?

<details>
<summary>Click to reveal answer</summary>

**A:** About 5 × 10⁻⁷ meters (500 nanometers)
</details>

---
## Card 11
**Q:** Why does the Earth's rotation make a poor time standard?

<details>
<summary>Click to reveal answer</summary>

**A:** The Earth's rotation is slowing down due to tidal friction, so days are getting longer over centuries.
</details>

---
## Card 12
**Q:** What is the distance to the edge of the observable universe?

<details>
<summary>Click to reveal answer</summary>

**A:** About 10²⁶ meters (or ~10 billion light-years in Feynman's time; now known to be ~46 billion light-years)
</details>

---
## Card 13
**Q:** How do we measure nuclear sizes?

<details>
<summary>Click to reveal answer</summary>

**A:** By firing high-energy particles through matter and measuring the effective cross section (scattering probability).
</details>

---
## Card 14
**Q:** What does the time-energy uncertainty relation say?

<details>
<summary>Click to reveal answer</summary>

**A:** Δt · ΔE ≥ ℏ/2. To know *when* something happened precisely, you must know less about *what* happened (energy).
</details>

---
## Card 15
**Q:** What is the lifetime of a π⁰ meson?

<details>
<summary>Click to reveal answer</summary>

**A:** About 10⁻¹⁶ seconds
</details>

---
## Card 16
**Q:** How did Feynman measure the distance to Venus?

<details>
<summary>Click to reveal answer</summary>

**A:** Radar: send a pulse, measure round-trip time, use distance = (c × t) / 2.
</details>

---
## Card 17
**Q:** What is Norman Ramsey's contribution to timekeeping?

<details>
<summary>Click to reveal answer</summary>

**A:** He built a hydrogen atomic clock in the 1960s, improving accuracy by 100× over earlier atomic clocks.
</details>

---
## Card 18
**Q:** What is the color-brightness method for measuring stellar distances?

<details>
<summary>Click to reveal answer</summary>

**A:** Measure a star's color to infer its intrinsic brightness, then use apparent brightness to calculate distance (inverse-square law).
</details>

---
## Card 19
**Q:** What is the gap factor between atomic and nuclear scales?

<details>
<summary>Click to reveal answer</summary>

**A:** About 10⁵ (100,000). Atoms are ~10⁻¹⁰ m, nuclei are ~10⁻¹⁵ m.
</details>

---
## Card 20
**Q:** Why can't we see objects smaller than the wavelength of light?

<details>
<summary>Click to reveal answer</summary>

**A:** Light waves diffract around small objects. You can't resolve details smaller than the wavelength you're using.
</details>
