## Section 2-4: Nuclei and Particles

This section explores the structure of the atomic nucleus, the discovery of new particles (mesons, baryons), and the four fundamental forces known to physics.

### ‚öõÔ∏è What Holds the Nucleus Together?
- **The Puzzle:** Protons in the nucleus repel each other electrically. What keeps them from flying apart?
- **The Answer:** A new force‚Äîthe **Strong Nuclear Force**.
- **Yukawa‚Äôs Idea (1935):** Just as electricity involves exchanging photons, the nuclear force might involve exchanging a new particle with mass.
- **Discovery:** The predicted particle (the **pion** or $\pi$-meson) was found in 1947. (The muon, found earlier, was a "wrong" particle that didn't fit the theory.)

### ü¶Å The Particle Zoo
- By the 1950s/60s, dozens of new particles were found in cosmic rays and accelerators.
- **Classification:**
  - **Baryons:** Heavy particles (Proton, Neutron, Lambda $\Lambda$, Sigma $\Sigma$...). Interaction strength: Strong.
  - **Mesons:** Intermediate mass (Pion $\pi$, Kaon $K$). Interaction strength: Strong.
  - **Leptons:** Light particles (Electron, Muon $\mu$, Neutrino $\nu$). Interaction strength: Weak (no strong force).
- **Strangeness:** A new quantum number ($S$) introduced to explain why some particles lived longer than expected.

### üí° Summary of our Knowledge (1963)
- **Outside the nucleus:** We understand almost everything (Quantum Electrodynamics).
- **Inside the nucleus:** We know the rules (Quantum Mechanics) but not the specific laws of force. The particle situation is "horrible" and complex.
- **The Goal:** To find the underlying simplicity behind this mess of particles (which later turned out to be the Standard Model and Quarks).
