# 5-2: Time
## What *Is* Time?
Feynman confronts one of the deepest questions in physicsâ€”and gives a surprisingly pragmatic answer.

## The Problem: Defining Time
Feynman begins with a philosophical puzzle:
> "What is time?"

**Webster's Dictionary:**
- "A time" = "a period"
- "A period" = "a time"

This is circular and unhelpful.

**Feynman's attempt:**
> "Time is what happens when nothing else happens."

Still not very useful!

**The honest answer:**
> "Maybe it is just as well if we face the fact that time is one of the things we probably cannot define (in the dictionary sense), and just say that it is what we already know it to be: **it is how long we wait!**"

This is classic Feynmanâ€”refusing to pretend we understand something we don't.

## What Matters: Measuring Time
> "What really matters anyway is not how we define time, but **how we measure it**."

Physics doesn't need a philosophical definition of timeâ€”it needs a **standard** for measurement.

**The key idea:** Use something that happens **periodically** (over and over again in a regular fashion).

**Example:** A day.
- The sun rises, sets, rises again.
- Days seem to repeat regularly.

## Are Days Really Periodic?
But waitâ€”are days *actually* the same length?

**Subjective experience:**
- Summer days feel longer than winter days.
- A boring day feels longer than an exciting one.

**The scientific question:**
> Can we *test* whether days are the same length?

**Feynman's method:** Compare one periodic phenomenon with another.

## The Hourglass Experiment
**Setup:**
- Use an hourglass as a second periodic clock.
- Have someone flip it every time the sand runs out (day and night).
- Count the number of "hours" (flips) from one morning to the next.

**Result:**
- The number of hours is **not the same each day**.

**Problem:** Which clock is wrong?
- The sun?
- The hourglass?
- Both?

We can't tell yet!

## Refining the Experiment: Noon to Noon
**Better idea:** Count hours from **noon to noon**.
- Noon = the instant when the sun is at its highest point in the sky.

**Result:**
- Now the number of hours is **the same** each day!

**Conclusion:**
- Both the "hour" (hourglass) and the "day" (noon to noon) appear to be regular.
- We have found two clocks that **agree**.

## The Limits of Verification
**Feynman's caution:**
> We have not *proven* that either clock is "really" periodic.

**Skeptical objection:**
> What if some omnipotent being slows down the sand every night and speeds it up during the day?

**Feynman's answer:**
> Our experiment doesn't answer this kind of question. All we can say is that we find that **a regularity of one kind fits together with a regularity of another kind**.

**The operational definition:**
> We base our definition of time on the **repetition of some apparently periodic event**.

This is the foundation of all timekeeping.
