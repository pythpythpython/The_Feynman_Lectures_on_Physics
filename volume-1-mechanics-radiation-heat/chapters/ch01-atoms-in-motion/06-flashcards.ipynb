# Volume 1, Chapter 1 – Atoms in Motion – Flashcards


This notebook provides flippable flashcards using `ipywidgets`.


```
import random
from dataclasses import dataclass
from IPython.display import display, Markdown
import ipywidgets as widgets
```

```
@dataclass
class Card:
    front: str
    back: str

cards = [
    Card(front=r'State Newton\'s second law.', back=r'$\\vec F = m\\vec a$'),
    # Add more cards here
]
```

```
current_index = 0
showing_front = True

front_button = widgets.Button(description='Show Front', button_style='primary')
back_button = widgets.Button(description='Show Back')
next_button = widgets.Button(description='Next Card')

output = widgets.Output()

def show_front(idx):
    with output:
        output.clear_output()
        display(Markdown('**Front**:\n\n' + cards[idx].front))

def show_back(idx):
    with output:
        output.clear_output()
        display(Markdown('**Back**:\n\n' + cards[idx].back))

def on_front_click(b):
    global showing_front
    showing_front = True
    show_front(current_index)

def on_back_click(b):
    global showing_front
    showing_front = False
    show_back(current_index)

def on_next_click(b):
    global current_index, showing_front
    current_index = random.randrange(len(cards))
    showing_front = True
    show_front(current_index)

front_button.on_click(on_front_click)
back_button.on_click(on_back_click)
next_button.on_click(on_next_click)

display(widgets.HBox([front_button, back_button, next_button]))
display(output)

# Show first card initially
show_front(current_index)
```
