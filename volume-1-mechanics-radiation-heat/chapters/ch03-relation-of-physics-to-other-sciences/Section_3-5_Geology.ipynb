# 3-5: Geology
## The Paradox of the Earth
Feynman presents a surprising fact: **We understand the interior of the sun far better than the interior of the earth.**

### 1. Meteorology (The Weather)
- **The Physicist's Problem:** We know the equations of air motion perfectly. Why can't we predict the weather?
- **Instability:** The air is swirling and twisting. It is a system of **Turbulent Flow**.
- **The Dam Analogy:** Water flows smoothly over a dam, but then breaks into chaotic blobs and drops. We cannot predict exactly where a specific drop will form. The same applies to air swirling over a mountain.

### 2. Mountain Building
- **Erosion:** We understand how wind and water wear mountains down.
- **Uplift:** But mountains are not getting lower on average. Something is pushing them up!
- **The Mystery:** What pushes? The theory is **convection currents** inside the Earth (hot rock rising, cool rock sinking). But the details of this "volcanism" are still largely mysterious (at least in Feynman's time).

### 3. The Core of the Earth
Why is the Earth harder to understand than a Star?
- **Stars:** Made of gas. The physics of gas at high pressure is relatively simple (Statistical Mechanics).
- **Earth:** Made of rock/iron. The physics of solids at high pressure is **extremely difficult**.
- **Mathematics:** We can't easily calculate how "squishy" or "brittle" rock is at 5,000 degrees and millions of atmospheres of pressure.

> **Key Lesson:** Complexity (Earth) is often harder to solve than Simplicity (Stars), even if the object is right beneath our feet.
