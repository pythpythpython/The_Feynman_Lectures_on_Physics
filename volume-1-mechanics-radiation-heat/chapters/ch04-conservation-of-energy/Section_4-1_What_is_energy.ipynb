# 4-1: What is energy?
The law of conservation of energy is among the deepest and most abstract in all of physics. Feynman illustrates its meaning by analogy, not by definition.

## The Law of Conservation
- There is a certain quantity, **energy**, that never changes in all natural phenomena so far observed.
- The law is not about a physical thing, but about a number that remains constant, no matter how complex the changes.
- We do NOT know what energy 'is.' Physics provides formulas, not a picture.

## Dennis the Menace Analogy
Imagine a child with 28 identical, indestructible blocks.

No matter what Dennis does, if we count carefully, there are always 28 blocksâ€”unless one:
- Is under the rug
- Out the window
- In a box (can't open)
- Inside the dirty bathtub (can't see)
So, Dennis' mom invents a *formula*:
- She weighs the box, measures the water, checks everywhere she can, and adds terms to her formula for hidden places.
- This formula always gives the same total: 
    
    $$
    \text{(blocks seen)} + \frac{\text{weight of box} - 16}{3} + \frac{\text{height of water} - 6}{1/4} + \ldots = \text{constant}
    $$

## What does this teach us about energy?
- **Energy is not a tangible object.** Just like the total block count, it is calculated from many *different forms* and formulas.
- **Forms of energy:**
    - Gravitational
    - Kinetic
    - Heat
    - Elastic
    - Chemical
    - Electrical
    - Nuclear
    - Mass ($E = mc^2$)
- The formulas are mechanical: you plug in values, sum everything, and the result is always the sameâ€”if no energy enters or leaves the system.

> "It is important to realize that in physics today, we have no knowledge of what energy is."
We can't say energy consists of 'blobs' or particles. It is a number constructed from formulas, always constant for a closed system.
