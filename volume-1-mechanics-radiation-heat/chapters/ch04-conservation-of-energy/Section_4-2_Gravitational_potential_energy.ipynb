# 4-2: Gravitational Potential Energy
## Deriving \( mgh \) from Pure Logic
Feynman does not appeal to calculus or Newton's laws. Instead, he uses a thought experiment with **reversible machines** and the impossibility of **perpetual motion** to derive the formula for gravitational energy.

## The Fundamental Hypothesis
> *There is no such thing as perpetual motion.*

- **Perpetual motion** = A machine that can lift a weight and return to its original state, creating energy from nothing.
- **Reversible machine** = An ideal, frictionless machine that can run forwards or backwards with no energy loss.

We assume:
1. Perpetual motion machines do not exist.
2. Reversible machines represent the best possible efficiency.

## The Reversible Machine Thought Experiment
**Suppose:**
- Machine A (reversible): Lowers 1 lb by 1 ft, lifts 3 lb by distance \(X\).
- Machine B (unknown): Lowers 1 lb by 1 ft, lifts 3 lb by distance \(Y\).

**Claim:** \(Y \leq X\).

**Proof by Contradiction:**
If \(Y > X\):
1. Use Machine B to lower 1 lb by 1 ft â†’ lifts 3 lb to height \(Y\).
2. Lower the 3 lb weight from \(Y\) to \(X\) â†’ gain free energy.
3. Use Machine A (backwards) to lift 1 lb back up â†’ restores the system.
4. Result: We have created energy from nothing. **Contradiction!**

Thus, \(Y \leq X\). Reversible machines are the most efficient.

## Deriving \( mgh \) Using a Ball Rack
Feynman now uses a reversible ball-lifting machine to calculate \(X\) exactly (Fig. 4-2).

**Setup:**
- Start with 3 balls on shelves (spaced \(X\) apart) and 1 ball on the ground.
- The reversible machine lowers 1 ball by 1 foot and lifts the rack by \(X\).

**Process:**
1. Load 3 balls onto the rack.
2. Lower 1 ball by 1 ft â†’ rack rises by \(X\).
3. Unload balls back onto shelves.
4. Net effect: One ball has been lifted by \(3X\).

**Energy Balance:**
- If \(3X > 1\) ft, we can lower that ball and repeat â†’ perpetual motion!
- Thus, \(3X \leq 1\).
- Running the machine backwards proves \(3X \geq 1\).
- Therefore: \(3X = 1\) â†’ \(X = 1/3\) ft.

**Generalization:**
$$
\text{(Weight}_1 \times \text{Height}_1) = \text{(Weight}_2 \times \text{Height}_2)
$$
This gives us the formula:
$$
PE = mgh
$$

## Stevinus' Brilliant Shortcut: The Inclined Plane
Feynman shows how the same result can be derived using an inclined plane (Fig. 4-3).

**Problem:**
- A 1 lb weight sits on a 3-4-5 triangle incline.
- A pulley connects it to a hanging weight \(W\).
- What must \(W\) be to balance?

**Energy Balance:**
- If \(W\) drops 5 ft, the 1 lb weight rises 3 ft.
- Energy conserved: \(W \times 5 = 1 \times 3\).
- Thus: \(W = 3/5\) lb.

But Stevinus found an even more elegant proof using a **chain of beads** draped over the triangle (Fig. 4-4). The chain cannot move on its own, so the force ratio must be 3:5.

## The Principle of Virtual Work
Feynman ends with a powerful generalization:
> *To solve statics problems, imagine the structure moves a tiny bit and apply energy conservation.*

**Example: The Weighted Rod (Fig. 4-6)**
- An 8 ft rod is supported at one end.
- A 60 lb weight is at the center (4 ft).
- A 100 lb weight is 2 ft from the support.
- How much weight \(W\) is needed at the far end to balance?

**Solution:**
Imagine \(W\) drops 4 inches.
- The center rises 2 inches.
- The 2 ft mark rises 1 inch.
Energy balance:
$$
-4W + 2(60) + 1(100) = 0
$$
$$
W = 55 \text{ lb}
$$
