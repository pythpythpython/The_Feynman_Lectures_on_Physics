# 3-2: Chemistry
## "Theoretical Chemistry is in fact Physics"
Feynman argues that the boundary between physics and chemistry has largely dissolved. 

### 1. The Reduction of Inorganic Chemistry
- **The Periodic Table:** Mendeleev organized the elements by their properties (rules of combination). 
- **The Explanation:** Quantum Mechanics later explained *why* these rules exist. The electron shells and energy levels determines how atoms bond.
- **The Limitation:** "We have knowledge of the rules, but we cannot play very well." 
  - We can write the SchrÃ¶dinger equation for any molecule.
  - We can *solve* it precisely only for the simplest ones (like Hydrogen).
  - For complex molecules, the math is too hard, so we still need experimental chemistry.

### 2. Statistical Mechanics: The Science of Heat
How do we predict the behavior of a gas if we can't track every single atom?

- **The Problem:** Trillions of atoms are colliding randomly. No computer can track them all.
- **The Solution:** We use *statistics* (averages/probabilities).
- **The Result:** Concepts like **Pressure**, **Temperature**, and **Entropy** emerge from these averages. This field is called **Thermodynamics** (or Statistical Mechanics).

> **Key Insight:** Heat is not a fluid; it is the random jiggling motion of atoms.

### 3. Organic Chemistry: The Bridge to Life
Historically, people thought living things contained a special substance. Feynman clarifies:

- **No Vital Force:** Organic molecules are made of the *exact same atoms* as inorganic ones, following the *exact same physical laws*.
- **Complexity:** The only difference is the arrangement. Organic molecules are larger and more complex.
- **Synthesis:** We can now make these "substances of life" in the lab (synthesizing urea, etc.), proving they are just physics/chemistry.
- **The Path:** Organic Chemistry $\to$ Biochemistry $\to$ Molecular Biology.
