**Section:** 5-7: Short Distances

**Question:**
> Why can't we see atoms with a regular microscope?

**AI Response:**
> Atoms are smaller than the wavelength of visible light (~500 nm). Light diffracts around objects smaller than its wavelength, so you can't resolve atomic-scale details (~0.1 nm) with visible light.

**Follow-up Thoughts:**
> This makes sense! So electron microscopes work because electrons have much shorter wavelengths. Cool connection to wave-particle duality.
